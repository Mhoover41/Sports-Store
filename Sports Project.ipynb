# 1.  KPI's:
- Profit
- Total Revenue
- Profit Margin
- Total number of orders

```python
q = """
SELECT 
    SUM(revenue) AS total_revenue,
    SUM(profit) AS total_profit,
    COUNT(*) as total_orders,
    (SUM(profit)/SUM(revenue)) * 100 AS profit_margin
FROM orders;
"""
pd.read_sql(q, engine)
```

# 2. Now I'd like to query the KPI's above by sport and in decending order of profit margin. 

```python
q = """
SELECT 
    Sport,
    ROUND(SUM(revenue),2) AS total_revenue,
    ROUND(SUM(profit),2) AS total_profit,
    COUNT(*) as total_orders,
    (SUM(profit)/SUM(revenue)) * 100 AS profit_margin
FROM orders
GROUP BY Sport
ORDER BY profit_margin DESC;
"""
pd.read_sql(q, engine)
```

# 3. How many customers gave a rating?  What was the average rating overall?

- These are rated 1-5.  1 is poor service and 5 being excellent customer service

```python
q = """
SELECT 
    (SELECT COUNT(*) FROM orders WHERE rating is NOT NULL) as number_of_ratings,
    ROUND(AVG(rating),2) AS average_rating
FROM orders;
"""
pd.read_sql(q, engine)
```

1193/2847=42% of customers provided a rating.  Average rating was 3.13

# 4.  Order and Group by each rating and show it's revenue, profit, and profit margin.

```python
q = """
SELECT 
    rating,
    SUM(revenue) as total_revenue,
    SUM(profit) as total_profit,
    ROUND((SUM(profit)/SUM(revenue)) * 100, 2) as profit_margin
FROM orders
WHERE rating is not null
GROUP BY rating
ORDER BY rating DESC;
"""
pd.read_sql(q, engine)
```

- Customers who gave a 3 rating had the highest profit margin. 

# 5.  Find the revenue, profit, and profit margin by State

```python
q = """
SELECT 
    c.state,
    SUM(o.revenue) as total_revenue,
    SUM(o.profit) as total_profit,
    ROUND((SUM(o.profit)/SUM(o.revenue)) * 100, 2) as profit_margin
FROM orders o
join customers c
on o.customer_id = c.customer_id
group by c.state
order by total_profit DESC;
"""
pd.read_sql(q, engine)
```

- California had the highest profits with Main having the lowest profit.

# 6.  Rank the total revenue, total profit, and profit margin

```python
q = """
SELECT 
    c.state,
    row_number() over (order by SUM(o.revenue) DESC) as revenue_rank,
    SUM(o.revenue) as total_revenue,
    row_number() over (order by SUM(o.profit) DESC) as profit_rank,
    SUM(o.profit) as total_profit,
    row_number() over (order by SUM(o.profit)/SUM(o.revenue) * 100 DESC) as margin_rank,
    ROUND((SUM(o.profit)/SUM(o.revenue)) * 100, 2) as profit_margin
FROM orders o
JOIN customers c
ON o.customer_id = c.customer_id
GROUP BY c.state
ORDER BY revenue_rank;
"""
pd.read_sql(q, engine)
```

- Maine had the lowest margin rank and Utah had the highest margin rank
- Maine had the lowest profit rank and CA had the highest profit rank
- Maine had the lowest revenue rank and CA had the highest revenue rank

# 7. Monthly Profits

```python
q = """
WITH monthly_profit AS (
    SELECT 
        MONTH(date) as month,
        SUM(profit) as total_profit
    FROM 
        orders
    GROUP BY MONTH(date)
)
SELECT 
    month,
    total_profit,
    LAG(total_profit) OVER (ORDER BY month) as previous_month_profit,
    ROUND(total_profit - LAG(total_profit) OVER (ORDER BY month),2) as profit_difference
FROM monthly_profit
ORDER BY month;
"""
pd.read_sql(q, engine)
```
